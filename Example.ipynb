```python
from MovingAverage import MovingAverage

from random import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# We generate an example sequence using random numbers
example_sequence = [random() for i in range(1000)];

'''
We initialize the moving average system using a sequence of 500 samples. This initializes and computes the ma vector inside the function as well
as the buffer and its pointer.
'''
win_size= 10;
ma_filter = MovingAverage(win_size);
ma_filter.compute_start(example_sequence[0:500]);
half_sequence = ma_filter.give_ma();


'''
Once the system is initialized, we use the system sample by sample to estimate their current moving average.
'''
for i in range(500,len(example_sequence)):
    ma_filter.compute_online(example_sequence[i])
full_sequence = ma_filter.give_ma();

# We plot the results
plt.figure(figsize=(10,10))
plt.subplot(211)
plt.plot(example_sequence, label='Input sequence')
plt.plot(half_sequence, label='Half signal smoothed')
plt.legend()
plt.subplot(212)
plt.plot(example_sequence, label='Input sequence')
plt.plot(full_sequence,label='Full signal smoothed')
plt.legend()
```
